```python
from cv2 import cv2
import numpy as np
import matplotlib.pyplot as plt
import copy
```

```python
%matplotlib notebook
```

```python
fig=plt.figure(1)
src=cv2.imread('o.tif')
src=src[0:700,:,0]

plt.subplot(1,2,1)
plt.imshow(src,cmap=plt.cm.gray)

plt.subplot(1,2,2)
plt.imshow(src,cmap=plt.cm.jet)

```

```python
fig=plt.figure(2)
src=cv2.blur(src,(9,9))
x = cv2.Sobel(src,cv2.CV_16S,1,0)
y = cv2.Sobel(src,cv2.CV_16S,0,1)
absX = cv2.convertScaleAbs(x)
absY = cv2.convertScaleAbs(y)
gradModel = cv2.addWeighted(absX,0.5,absY,0.5,0)



plt.imshow(gradModel,cmap=plt.cm.gray)
```

```python
fig=plt.figure(3)
clahe=cv2.createCLAHE(clipLimit=2.0, tileGridSize=(20,20))
claheSameHis=np.zeros(gradModel.shape,dtype=np.uint8)
claheSameHis[:,:]=clahe.apply(gradModel[:,:])

plt.imshow(claheSameHis,cmap=plt.cm.gray)
```

```python
fig=plt.figure(4)

__,threImg=cv2.threshold(gradModel,5,255,cv2.THRESH_BINARY)
morImg=cv2.morphologyEx(threImg,cv2.MORPH_OPEN,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(2,2)))
plt.imshow(threImg,cmap=plt.cm.gray)
```

```python
fig=plt.figure(5)

lines = cv2.HoughLinesP(morImg, 1, np.pi / 180, 100, minLineLength=50, maxLineGap=2)
showImg=cv2.imread('o.tif')
showImg=showImg[0:700,:,:]
for line in lines:
    x1, y1, x2, y2 = line[0]

    cv2.line(showImg, (x1, y1), (x2, y2), (0, 255, 0), 1)
    

plt.imshow(showImg,cmap=plt.cm.gray)
```

```python
fig=plt.figure(6)
a_values=list()
b_values=list()
for line in lines:
    x1, y1, x2, y2 = line[0]
    a=(y1-y2)/((x1-x2)+0.000001)
    
    A=y2-y1
    B=x1-x2
    C=x2*y1-x1*y2
    sig=(-C/(B+0.000001))/np.abs((-C/(B+0.000001)))
    distance = np.abs(C / np.sqrt(A*A + B*B))*sig
    
    b=y1-a*x1
    a=np.arctan(a)
    a_values.append(a)
    b_values.append(distance)

a_values=np.array(a_values)    
b_values=np.array(b_values)
maxA=np.max(np.abs(a_values))
maxB=np.max(np.abs(b_values))
a_values=a_values/maxA
b_values=b_values/maxB
plt.scatter(x=a_values,y=b_values)
plt.show()
```

```python
fig=plt.figure(7)
cluster=np.zeros(shape=(600,600))
for index,__ in enumerate(a_values):
    a=int((a_values[index]+1)*250+50)
    b=int((b_values[index]+1)*250+50)
    cv2.circle(cluster,(a,b),3,(255,255,255),-1)
    
plt.imshow(cluster)

```

```python
fig=plt.figure(8)
core=np.ones(shape=(1,30))
core[0,10:20]=0
core=core/np.sum(core)
showcluster=cv2.filter2D(cluster,-1,kernel=core.T)
__,showcluster=cv2.threshold(showcluster,150,255,cv2.THRESH_BINARY)
plt.imshow(showcluster)


```

```python
fig=plt.figure(9)
newcluster=cv2.morphologyEx(showcluster,cv2.MORPH_OPEN,cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(1,1)))
__,contours,__=cv2.findContours(newcluster.astype(np.uint8),cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
new_a_values=list()
new_b_values=list()
centerImg=cv2.imread('o.tif')
centerImg=centerImg[0:700,:,:]

for contour in contours:
    M = cv2.moments(contour)
    if M["m00"]!=0:
        x = int(M["m10"] / M["m00"])
        y = int(M["m01"] / M["m00"])
        a=((x-50)/250-1)
        b=((y-50)/250-1)*maxB

        new_a_values.append(a)
        new_b_values.append(b)

        if a!=0 and a!=1 and a!=-1:
            a=a*maxA
            point0=(0,int(b/np.cos(a)))
            point1=(src.shape[1],int(src.shape[1]*np.tan(a)+b/np.cos(a)))

        else:
            if a==0:
                point0=(0,int(np.abs(b)))
                point1=(src.shape[1],int(np.abs(b)))
            else:
                point0=(int(np.abs(b)),0)
                point1=(int(np.abs(b)),src.shape[0])

        #centerImg=cv2.line(centerImg,(point0[1],point0[0]),(point1[1],point1[0]),(0,255,0),1)
        centerImg=cv2.line(centerImg,point0,point1,(0,255,0),2)
plt.imshow(centerImg) 

```
